```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rf_rmse = 12.5      # your value
rmse_lstm = 10.2   # your value
rmse_gru = 10.8    # your value
rmse_tf = 9.7      # your value
```

```python
import pandas as pd

results_df = pd.DataFrame({
    "Model": ["Random Forest", "LSTM", "GRU", "Transformer"],
    "RMSE": [rf_rmse, rmse_lstm, rmse_gru, rmse_tf]
})

results_df
```

```python
results_df.to_csv("final_results.csv", index=False)

print("CSV file saved successfully")
```

```python
plt.figure(figsize=(8,5))

sns.barplot(x="Model", y="RMSE", data=results_df)

plt.title("Model Comparison (RMSE)")
plt.ylabel("Error")

for i, v in enumerate(results_df["RMSE"]):
    plt.text(i, v, str(round(v,2)), ha='center')

plt.savefig("model_comparison.png")   # ✅ ADD THIS
plt.show()
```
